```python
import numpy as np
from telebot import TeleBot
import warnings
warnings.filterwarnings("ignore")
from __future__ import print_function
BOT_TOKEN = "TOKEN"
bot=TeleBot(BOT_TOKEN)
import gate_api
configuration = gate_api.Configuration(host = "https://api.gateio.ws/api/v4")
api_client = gate_api.ApiClient(configuration)
from gate_api.exceptions import ApiException, GateApiException
@bot.message_handler(commands=['start'])
def start_ex(message):
     
     
    bot.send_message(message.chat.id,  'this bot created to give you the last Stats of Spot and Margin Data, Please Write   margin eth  or  sport btc  to see how it works' )





def Margin(x,y):
    api_instance = gate_api.MarginApi(api_client)
    currency_pair = x + '_USDT' 

     
    
    try:
    # Query one single margin currency pair
         api_response = api_instance.get_margin_currency_pair(currency_pair)
         return(api_response)
    except GateApiException as ex:
         print("Gate api exception, label: %s, message: %s\n" % (ex.label, ex.message))
    except ApiException as e:
         print("Exception when calling MarginApi->get_margin_currency_pair: %s\n" % e)
   
        
def Spot(x,y):
    
    
    api_instance = gate_api.SpotApi(api_client)
    currency_pair = x + '_'+ y  
     
    api_response = api_instance.list_tickers(currency_pair=currency_pair)
    data=api_response
    
    
      
    return(data) 
 
@bot.message_handler(content_types=['text'])
def crypto_price_message_handler(message):
    A = (message.text).upper()
    
    d=len(A.split())
    for i in range (d):
        function1=A.split(' ')[0]
        variable2=A.split(' ')[-2] 
        variable1=A.split(' ')[-1]
         
    if (function1 == 'MARGIN'):
        B=Margin(variable1)
         
        bot.send_message(chat_id=message.chat.id, text= B   )
    elif (function1 == 'SPOT'):
        B=Spot(variable2,variable1)
           
        
        bot.send_message(chat_id=message.chat.id, text= ' '+ B[0].currency_pair+' '+"price :("+B[0].last+') '+'\n'+"Change is"+' ('+B[0].change_percentage+"%)"+'\n'+"Trading Vol:("+' '+B[0].quote_volume+'$)') 
    else :
        bot.send_message(chat_id=message.chat.id, text="please type following structure margin btc or spot btc")    
    
if __name__ == '__main__':
    bot.infinity_polling()
```
